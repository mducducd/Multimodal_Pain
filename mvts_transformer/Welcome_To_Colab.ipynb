```python
def interpolate_missing(y):
    """
    Replaces NaN values in pd.Series `y` using linear interpolation
    """
    if y.isna().any():
        y = y.interpolate(method='linear', limit_direction='both')
    return y
# Replace NaN values
grp = df.groupby(by=df.index)
df = grp.transform(interpolate_missing)
```
