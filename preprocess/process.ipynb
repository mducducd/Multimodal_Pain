```python
import torch
import torch.nn as nn
import torchvision.models as models

# Custom VGG3D model based on VGG16
class VGG3D(nn.Module):
    def __init__(self, num_classes=3, pretrained=True):
        super(VGG3D, self).__init__()
        
        # Load the VGG16 model
        vgg2d = models.vgg16(pretrained=pretrained)
        
        # Modify the features part to handle 3D input
        self.features = nn.Sequential(
            nn.Conv3d(3, 64, kernel_size=(3, 3, 3), padding=(1, 1, 1)), # First Conv3d layer
            nn.ReLU(inplace=True),
            nn.MaxPool3d(kernel_size=(1, 2, 2), stride=(1, 2, 2)), # MaxPool3d
            
            nn.Conv3d(64, 128, kernel_size=(3, 3, 3), padding=(1, 1, 1)),
            nn.ReLU(inplace=True),
            nn.MaxPool3d(kernel_size=(2, 2, 2), stride=(2, 2, 2)),
            
            nn.Conv3d(128, 256, kernel_size=(3, 3, 3), padding=(1, 1, 1)),
            nn.ReLU(inplace=True),
            nn.Conv3d(256, 256, kernel_size=(3, 3, 3), padding=(1, 1, 1)),
            nn.ReLU(inplace=True),
            nn.MaxPool3d(kernel_size=(2, 2, 2), stride=(2, 2, 2)),
            
            nn.Conv3d(256, 512, kernel_size=(3, 3, 3), padding=(1, 1, 1)),
            nn.ReLU(inplace=True),
            nn.Conv3d(512, 512, kernel_size=(3, 3, 3), padding=(1, 1, 1)),
            nn.ReLU(inplace=True),
            nn.MaxPool3d(kernel_size=(2, 2, 2), stride=(2, 2, 2)),
            
            nn.Conv3d(512, 512, kernel_size=(3, 3, 3), padding=(1, 1, 1)),
            nn.ReLU(inplace=True),
            nn.Conv3d(512, 512, kernel_size=(3, 3, 3), padding=(1, 1, 1)),
            nn.ReLU(inplace=True),
            nn.MaxPool3d(kernel_size=(2, 2, 2), stride=(2, 2, 2))
        )
        
        # Modify the classifier part
        self.classifier = nn.Sequential(
            nn.Linear(512 * 7 * 7 * 1, 4096),  # Adjust based on flattened size
            nn.ReLU(True),
            nn.Dropout(),
            nn.Linear(4096, 4096),
            nn.ReLU(True),
            nn.Dropout(),
            nn.Linear(4096, num_classes),
        )

    def forward(self, x):
        x = self.features(x)  # Pass through the modified VGG layers
        x = x.view(x.size(0), -1)  # Flatten
        x = self.classifier(x)  # Fully connected layers
        return x

# Example usage
if __name__ == "__main__":
    # Input shape: [batch_size=16, channels=3, frames=16, height=224, width=224]
    input_tensor = torch.randn(16, 3, 16, 224, 224)

    # Initialize the VGG3D model for 3-class classification
    model = VGG3D(num_classes=3, pretrained=False)

    # Forward pass
    output = model(input_tensor)
    print(output.shape)  # Expected output: [16, 3]
```

```python
import cv2
import dlib
import numpy as np

# Load dlib's pre-trained face detector and facial landmark predictor
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor('/home/duke/Workspace/MARLIN/ckpt/shape_predictor_68_face_landmarks.dat')

def get_landmarks(image):
    """Detect facial landmarks using dlib"""
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    faces = detector(gray)
    if len(faces) > 0:
        landmarks = predictor(gray, faces[0])
        return np.array([[p.x, p.y] for p in landmarks.parts()])
    return None

def calculate_landmark_distance(landmarks1, landmarks2):
    """Calculate Euclidean distance between two sets of landmarks"""
    return np.linalg.norm(landmarks1 - landmarks2)

def keyframe_selection_landmarks(video_path, num_keyframes=16):
    """Select keyframes based on changes in facial landmarks"""
    cap = cv2.VideoCapture(video_path)
    
    if not cap.isOpened():
        print("Error: Could not open video.")
        return

    prev_landmarks = None
    keyframes = []
    frame_id = 0
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    selected_frame_ids = []

    while frame_id < total_frames:
        ret, frame = cap.read()
        if not ret:
            break
        
        # Get facial landmarks for the current frame
        landmarks = get_landmarks(frame)
        if landmarks is not None:
            if prev_landmarks is None:
                # Always include the first frame
                keyframes.append((frame_id, frame))
                selected_frame_ids.append(frame_id)
            else:
                # Calculate landmark distance from the previous frame
                landmark_dist = calculate_landmark_distance(landmarks, prev_landmarks)
                if landmark_dist > 0.0:  # Threshold for significant change
                    keyframes.append((frame_id, frame))
                    selected_frame_ids.append(frame_id)
            
            prev_landmarks = landmarks

        # Uniformly sample by skipping frames to ensure selection of 16 keyframes
        frame_id += total_frames // num_keyframes

    cap.release()
    
    # Limit to the desired number of keyframes (if too many were selected)
    keyframes = keyframes[:num_keyframes]
    return keyframes

# Usage example
video_path = '/home/duke/Workspace/BioVid-A/Validation/cropped/082414_m_64/082414_m_64-PA4-043.mp4'
keyframes = keyframe_selection_landmarks(video_path, num_keyframes=16)

# Saving keyframes as images
for idx, frame in keyframes:
    cv2.imwrite(f'keyframe_{idx}.jpg', frame)

print(f"Selected {len(keyframes)} keyframes.")
```

```python
import cv2
import dlib
import numpy as np
import glob
import os
# Load dlib's pre-trained face detector and facial landmark predictor
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor('/home/duke/Workspace/MARLIN/ckpt/shape_predictor_68_face_landmarks.dat')

def get_landmarks(image):
    """Detect facial landmarks using dlib"""
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    faces = detector(gray)
    if len(faces) > 0:
        landmarks = predictor(gray, faces[0])
        return np.array([[p.x, p.y] for p in landmarks.parts()])
    return None

def calculate_landmark_distance(landmarks1, landmarks2):
    """Calculate Euclidean distance between two sets of landmarks"""
    return np.linalg.norm(landmarks1 - landmarks2)

def keyframe_selection_auto_threshold(video_path, min_keyframes=16, max_keyframes=100, initial_threshold=20.0, step=2.0, save_path=None):
    """Auto-adjust threshold to ensure at least 16 keyframes are selected"""
    cap = cv2.VideoCapture(video_path)
    
    if not cap.isOpened():
        print("Error: Could not open video.")
        return

    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    keyframes = []
    
    threshold = initial_threshold  # Start with an initial high threshold
    prev_landmarks = None

    # Function to process frames with a given threshold
    def select_keyframes(threshold):
        cap.set(cv2.CAP_PROP_POS_FRAMES, 0)  # Reset video to start
        prev_landmarks = None
        selected_keyframes = []

        for frame_id in range(total_frames):
            ret, frame = cap.read()
            if not ret:
                break

            landmarks = get_landmarks(frame)
            if landmarks is not None:
                if prev_landmarks is None:
                    selected_keyframes.append((frame_id, frame))  # Always add the first frame
                else:
                    landmark_dist = calculate_landmark_distance(landmarks, prev_landmarks)
                    if landmark_dist > threshold:
                        selected_keyframes.append((frame_id, frame))
                prev_landmarks = landmarks

        return selected_keyframes
    keyframes = select_keyframes(threshold)
    # Iteratively adjust the threshold to ensure we have at least `min_keyframes`
    while len(keyframes) < min_keyframes and threshold > 0:
        keyframes = select_keyframes(threshold)
        threshold -= step  # Lower the threshold if fewer than 16 frames are selected

    # Ensure the number of keyframes is not greater than `max_keyframes`
    if len(keyframes) > max_keyframes:
        keyframes = keyframes[:max_keyframes]

    cap.release()
    print(save_path, len(keyframes))
    for idx, frame in keyframes:
        cv2.imwrite(save_path+'/keyframe_{}.jpg'.format(str(idx)), frame)
    
    return keyframes



# Usage example
# video_path = '/home/duke/Workspace/BioVid-A/Train/cropped/071309_w_21/071309_w_21-PA2-064.mp4'
# keyframes = keyframe_selection_landmarks(video_path, threshold=15.0)  # Increased threshold

# # Saving all keyframes as images
# for idx, frame in keyframes:
#     cv2.imwrite(f'keyframe_{idx}.jpg', frame)
path = '/home/duke/Workspace/AI4Pain_Dataset_full'


files = glob.glob(path + '/*/*/*/*.mp4')
total = len(files)
n=0
for file in files:
    if not os.path.exists(file.replace('.mp4','')):
        os.makedirs(file.replace('.mp4',''))
    print(file)
    keyframe_selection_auto_threshold(file, initial_threshold=15.0, step=4 ,save_path=file.replace('.mp4',''))
    n+=1
    print(n,'/',total)
```

```python
import os

# Path to the root directory containing all subjects' video directories
dataset_dir = "/home/duke/Workspace/BioVid-A/Train/cropped"
output_dir = "loso_splits"  # Directory to store the generated folders for each fold
subject_dirs = sorted(os.listdir(dataset_dir))

# Create output directory if it doesn't exist
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Loop over each subject to create corresponding train.txt and val.txt for LOSO
for val_subject in subject_dirs:
    train_videos = []
    val_videos = []

    # Iterate over all subjects to create training and validation splits
    for subject in subject_dirs:
        subject_path = os.path.join(dataset_dir, subject)
        video_files = sorted(os.listdir(subject_path))  # List all files for the subject
        
        # Filter to only include files with 'BL1' or 'PA4' in the name and with a '.mp4' extension
        filtered_videos = [video for video in video_files if ('BL1' in video or 'PA4' in video) and video.endswith('.mp4')]
        
        # Validation set: Videos from the left-out subject
        if subject == val_subject:
            for video_file in filtered_videos:
                # Remove the .mp4 extension
                video_path = os.path.join(subject_path, video_file.replace('.mp4', ''))
                val_videos.append(video_path)
        else:
            # Training set: Videos from all other subjects
            for video_file in filtered_videos:
                # Remove the .mp4 extension
                video_path = os.path.join(subject_path, video_file.replace('.mp4', ''))
                train_videos.append(video_path)

    # Create a new folder for this fold in the output directory
    fold_dir = os.path.join(output_dir, val_subject)
    os.makedirs(fold_dir, exist_ok=True)

    # Write train.txt and val.txt for the current validation subject
    train_txt_path = os.path.join(fold_dir, 'train.txt')
    val_txt_path = os.path.join(fold_dir, 'val.txt')

    # Save train.txt
    with open(train_txt_path, 'w') as train_file:
        for video in train_videos:
            train_file.write(video.replace('/home/duke/Workspace/BioVid-A/', '') + '\n')

    # Save val.txt
    with open(val_txt_path, 'w') as val_file:
        for video in val_videos:
            val_file.write(video.replace('/home/duke/Workspace/BioVid-A/', '') + '\n')

    print(f"Created train.txt and val.txt for {val_subject} in {fold_dir}")
```
